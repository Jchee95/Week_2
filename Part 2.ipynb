```python
import pandas as pd
import numpy as np
f4 = pd.read_csv("titanicSurvival_m.csv", encoding = 'latin-1')
count_nan = f4.isna().sum().sum()
print(count_nan)
avg = f4['Age'].mean()
med = f4['Age'].median()
s_d = f4['Age'].std()
print('Mean of Age: ' + str(avg))
print('Median of Age: ' + str(med))
print('Standard deviation of Age: ' + str(s_d))
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
f4 = pd.read_csv("titanicSurvival_m.csv", encoding = 'latin-1')
df = pd.DataFrame(f4)
df[["Age", "Fare"]] = df[["Age", "Fare"]].fillna(0)
fig = plt.figure()
col_10 = df["Fare"]
col_6 = df["Age"]
plt.xlabel('Age')
plt.ylabel('Fare')
plt.scatter(col_6,col_10, s = 10, color = 'red')
plt.show()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
f4 = pd.read_csv("titanicSurvival_m.csv", encoding = 'latin-1')
df = pd.DataFrame(f4)
avg_1 = f4['Age'].mean()
avg_2 = f4['Fare'].mean()
df["Age"] = df["Age"].fillna(avg_1)
df["Fare"] = df["Fare"].fillna(avg_2)
fig = plt.figure()
col_10 = df["Fare"]
col_6 = df["Age"]
plt.xlabel('Age')
plt.ylabel('Fare')
plt.scatter(col_6,col_10, s = 10, color = 'red')
plt.show()
```
